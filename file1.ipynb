```python
# Convert the date column to datetime format.

df['date']=pd.to_datetime(df['date'],format='%Y-%m-%d')
df
```

```python
# Show unique values in phase column.

df['phase'].unique()
```

```python
# Count unique values in phase column.

df['phase'].nunique()
```

```python
# Create a new column: price_spread = brent - wti.

df['price_spread']=df['brent_usd_barrel']-df['wti_usd_barrel']
df
```

```python
# Calculate mean Brent oil price.

df[['brent_usd_barrel','wti_usd_barrel','dubai_usd_barrel']].mean()
```

```python
# Calculate maximum WTI price.

df[['wti_usd_barrel']].max()
```

```python
# Calculate minimum gas price.

df[['us_gas_avg_gallon']].min()
```

```python
# Count frequency of each phase.

df['phase'].value_counts()
```

```python
# Find the information of rows where gas price less than 3.

df[df['us_gas_avg_gallon']<3]
```

```python
# Find the information of rows where Brent price is between  80 and 100.

df[(df['brent_usd_barrel']>=80) & (df['brent_usd_barrel']<=100)]
```

```python
# Sort dataset by price spread

df.sort_values(by='price_spread')
```

```python
# Extract data for first 10 war days

df[df['war_day']<=10]
```

```python
# Calculate average Oil price across phase.

df.groupby('phase')[['brent_usd_barrel']].mean()
```

```python
# Find max and min oil prices per phase.

df.groupby('phase')['brent_usd_barrel'].agg(['max','min'])
```

```python
# Aggregate multiple columns using .agg()

df.groupby('phase').agg({'brent_usd_barrel':['max','min','mean'],'wti_usd_barrel':['max','min','mean','count'],
'dubai_usd_barrel':['max','min','mean','sum']})
```

```python
# Show relation among Brent, WTI and Dubai Oil Prices Over Time.

plt.plot(df.index,df['brent_usd_barrel'],label='Brent')
plt.plot(df.index,df['wti_usd_barrel'],label='WTI')
plt.plot(df.index,df['dubai_usd_barrel'],label='Dubai')
plt.title('Oil price over time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Show visualization of Average Brent Prices per Phase.

avg_brent=df.groupby('phase')['brent_usd_barrel'].mean()
avg_brent.plot(kind='bar')
plt.title('Average Brent Price per Phase')
plt.xlabel('Phase')
plt.ylabel('Brent Price')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Plots the  Oil and Fuel Prices on same graph.

fig,axes=plt.subplots(1,2)
# Oil prices
axes[0].plot(df.index,df['brent_usd_barrel'],label='Brent')
axes[0].plot(df.index,df['wti_usd_barrel'],label='WTI')
axes[0].set_title('Oil Prices')
axes[0].legend()
axes[0].grid()

# Fuel prices
axes[1].plot(df.index,df['us_gas_avg_gallon'],label='Gas')
axes[1].plot(df.index,df['us_diesel_avg_gallon'],label='Diesel')
axes[1].set_title('Fuel Prices')
axes[1].legend()
axes[1].grid()
plt.show()
```

```python
# Show the visualization of Iran Oil production.

plt.hist(df['iran_production_mbpd'],bins=10)
plt.title('Distribution of Iran Oil Production')
plt.xlabel('Production(mbpd)')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Show the distribution of caterogy column phase.

df['phase'].value_counts()
```

```python
# Show the Distribution of Phases.

phase_counts=df['phase'].value_counts()
plt.pie(phase_counts,labels=phase_counts.index,autopct='%1.1f%%')
plt.title('Distribution of War Phases')
plt.show()
```

```python
# Show Brent Price trend by war phase.

sns.lineplot(x=df.index, y='brent_usd_barrel',hue='phase',data=df)
plt.title("Brent Price Trend by War Phase")
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Total Ships Crossing Strait of Hormuz per Phase

df.groupby('phase')[['strait_hormuz_daily_ships']].max()
```

```python
# Show Total Ships Crossing Strait of Hormuz per Phase.

#Aggregation of total ships per phase
ships_per_phase=df.groupby('phase')['strait_hormuz_daily_ships'].max().reset_index()
# Plot
sns.barplot(x='phase',y='strait_hormuz_daily_ships',data=ships_per_phase)
plt.title('Total Ships Crossing Strait of Hormuz per Phase')
plt.xlabel('Phase')
plt.ylabel('Number of Ships')
plt.xticks(rotation=45)
plt.show()
```

```python
# To show Oil vs Gas Price with Phase & Ship Activity.

sns.scatterplot(x='brent_usd_barrel',y='us_gas_avg_gallon',hue='phase',style='phase',size='strait_hormuz_daily_ships',
data=df)
plt.title('Oil vs Gas Price with Phase & Ship Activity')
plt.xlabel('Brent Oil Price')
plt.ylabel('Gas Price')
plt.legend(bbox_to_anchor=(1.05,1),loc='upper left')
plt.grid()
# plt.show()
```

```python
# Create a plot visualize the distribution of brent_usd_barrel prices across different phase.

sns.catplot(x='phase',y='brent_usd_barrel',kind='box',hue='phase',data=df)
plt.xticks(rotation=45)
```

```python
# Create plot to visualize the distribution of Ships in Strait of Harmuz

sns.violinplot(df['strait_hormuz_daily_ships'])
```

```python
# Create a plot to visualize the distribution of diesel prices across different phase.

sns.catplot(x='phase',y='us_diesel_avg_gallon',kind='bar',hue='phase',data=df,height=6,aspect=1)
plt.xticks(rotation=45)
plt.show()
```
